```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# Read the CSV files
df1 = pd.read_csv('energy-4.csv')
df2 = pd.read_csv('energy-25.csv')

# Create a new figure
plt.figure(figsize=(12, 6))

# Plot the data from the first DataFrame
plt.plot(df1['Cumulative Package Energy_0 (Joules)'], label='Experiment October 4th, 2023')

# Plot the data from the second DataFrame
plt.plot(df2['Cumulative Package Energy_0 (Joules)'], label='Experiement October 25th, 2023')

plt.xlabel("Iteration")
plt.ylabel("Cumulative Package Energy (Joules) ")

plt.title("Cumulative Package Energy (Joules) for two experiments (October 4th and 25th, 2023)")
plt.legend(title="File")
plt.ylim(0,850)

plt.show()
```
